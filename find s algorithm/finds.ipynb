```python
# findS Algorithm
def expand(x,s):
    for i in range(len(s)):
        if x[i] != s[i] and s[i] == '': s[i] = x[i]
        elif x[i] != s[i] and s[i] != '': s[i] = '?'
    print(f's = {s}')
    print(x)
    return s

def finds(x,t):
    s = ['']*len(x[0])
    print(s)
    for i in range(len(x)):
        # print(x[i])
        print(i)
        if t[i] == 'Yes': s = expand(x[i],s)
    return s
```

```python
# Candidate Elimination Algorithm

def expand_x(x,s):
    for i in range(len(s)):
        if x[i] != s[i] and s[i] == '': s[i] = x[i]
        elif x[i] != s[i] and s[i] != '': s[i] = '?'
    # print(f's = {s}')
    # print(x)
    return s

def expand_g(x,s,g,_g):
    for i in range(len(s)):
        if x[i] != s[i] and s[i] != '?': 
            g.append(['?']*len(s))
            g[len(g)-1][i] = s[i]
            _g.append(s[i])
            
    return g
        
def check_g(x,g,_g): 
    l = []
    for i in range(len(_g)):
        if _g[i] not in x:
            print(f'i = {i}')
            l.append(_g[i])
            g.pop(i)
    for i in l: _g.remove(i)
    return g
            
    
def candidateElimination(x,t):
    s = ['']*len(x[0])
    g = [] ; result = [] ; _g = []
    for i in range(len(x)):
        # print(x[i])
        print(i)
        if t[i] == 'Yes': 
            s = expand_x(x[i],s)
            if len(g) > 0: g = check_g(x[i],g,_g)
        elif t[i] == 'No': g = expand_g(x[i],s,g,_g)
        print(f's = {s}')
        print(f'g = {g}')
        print(f'-g = {_g}')
    print(f'ss = {s}')
    print(f'gg = {g}')
    print(f'-gg = {_g}')
    for i in range(len(g)):
        for j in range(len(s)):
            if s[j] != g[i][j] : 
                l = g[i].copy()
                l[j] = s[j]
                # print(f'l = {l}')
                if l not in result:
                    result.append(l) 
    return result
```
