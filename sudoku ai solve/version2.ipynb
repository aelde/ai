```python
# _____________________ sudoku checker

def sub_box(ar, i, j):
    a = list(ar[i * 3:i * 3 + 3, j * 3:j * 3 + 3].flatten())
    b = check_3_ar(a)
    if b == True: return f'Box True'
    return f'Box False'

def sub_col(ar,n):
    # ar = ar[:,n]
    b = check_3_ar(ar[:,n])
    if b == True: return f'Col True'
    return f'Col False'

def sub_row(ar,n):
    # ar = ar[n,:]
    b = check_3_ar(ar[n,:])
    if b == True: return f'Row True'
    return f'Row False'

def check_3_ar(ar):
    ar = sorted(list(ar))
    a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    return ar == a

def auto_check_9(ar):
    a = {}
    k = 0
    for i in range(3):
        for j in range(3):
            print(i, j)
            b = sub_box(ar, i, j)
            a.update({b})
            c = sub_col(ar, k)
            a.update({c})
            d = sub_row(ar, k)
            a.update({d})
            print(a)
            print(b)
            print(c)
            print(d)
            k += 1
    if b == f'Box True' and c == f'Col True' and d == f'Row True': 
        print('True')
        return True
    else: 
        print('False')
        return False

            
```
